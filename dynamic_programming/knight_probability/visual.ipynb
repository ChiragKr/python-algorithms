```python
import knight_probability_v2 as kpv2
import numpy as np
import seaborn as sns
```

```python
n = 8
k = 2
probability = np.zeros(shape=(n,n), dtype=float)
for x in range(n):
    for y in range(n):
        probability[n-1-y][n-1-x] = kpv2.main(4, [n, k, x, y])
```

```python
probability
```

```python
ax = sns.heatmap(probability, vmin=0, vmax=1, linewidths=.5, annot=True)
```
